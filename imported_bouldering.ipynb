## Athlete stats

```python
athletes = events.get_athlete_data()

# single athlete data
tomoa = athletes["Tomoa Narasaki"]
tomoa.data
```

## Top ranked overall

```python
men, women = athletes.get_stats("Men"), athletes.get_stats("Women")
stats_dict = {"Men": men, "Women": women}

for sex, stats in stats_dict.items():
    display(HTML(f'<h3>{sex}</h3>'))
    display(stats.sort_values("ranking_B", ascending=False).head(10))
```

## Top ranked per event

```python
for sex, stats in stats_dict.items():
    display(HTML(f'<h3>{sex}</h3>'))
    display(stats.sort_values("ranking_per_event_B", ascending=False)
            .reset_index(drop=True).head(10)[["name", "country","ranking_per_event_B"]])
```

## Top ranked by top success

```python
for sex, stats in stats_dict.items():
    display(HTML(f'<h3>{sex}</h3>'))
    tops = stats.sort_values("top_percentage", ascending=False).reset_index(drop=True).head(10)
    tops["top_percentage"] = tops["top_percentage"].apply(lambda x: x*100).round(2)
    display(tops[["name", "country","tops","zones","top_percentage"]])
```

## Effect of height

```python
sex = "Men"

# filter out outliers
min_height = {"Men": 160, "Women": 140}
max_height = {"Men": 200, "Women": 180}
    
data = athletes.get_stats(sex)
data = data[(
            data.height > min_height[sex]) 
            & (data.height < max_height[sex]) 
            & (data.top_percentage > 0
              )][["name","height", "top_percentage"]]
data["top_percentage"] = data["top_percentage"].apply(lambda x: round(x, 2))*100

labels = {'height': "HEIGHT", 'top_percentage': "TOP %"}
fig = px.scatter(data, x="height", y="top_percentage", hover_name="name", 
                 labels=labels, trendline='ols', trendline_color_override=ifsc_pink,
                width=560, height=450)

fig.update_yaxes(range=(0,100))
fig.update_traces(marker=dict(color=ifsc_blue, size=10, line=dict(color=ifsc_blue_dark,width=2)))
fig.update_layout(margin=dict(l=50, r=100, t=30, b=20), title="Topping percentage vs. height",title_x=0.5, width=1000)
img = fig.show('png')
```

## Routesetting difficulty

```python
# get yearly mean top rate data 
years = [year for year in range(2008,2023)]
yearly_data = pd.DataFrame({"year": years})
avg_top_perc = {"men": [], "women": []}

for year in years:
    
    athletes = events.get_athlete_data(year)
    for sex in ["Men", "Women"]:

        stats = athletes.get_stats(sex)
        avg_top_perc[sex.lower()].append(stats["top_percentage"].mean())
        
for sex in ["men", "women"]:
    yearly_data[sex+"_avg_top_percentage"] = avg_top_perc[sex]

# plot results
fig = go.Figure()

fig.add_trace(go.Scatter(x=yearly_data["year"], y=yearly_data["men_avg_top_percentage"], 
                         line=dict(color=ifsc_blue, width=4),
                        marker=dict(size=8, line=dict(color=ifsc_blue_dark, width=2)), name="Men"))
fig.add_trace(go.Scatter(x=yearly_data["year"], y=yearly_data["women_avg_top_percentage"], 
                         line=dict(color=ifsc_pink, width=4),
                        marker=dict(size=8, line=dict(color=ifsc_pink_dark, width=2)), name="Women"))

fig.update_layout(margin=dict(l=20, r=20, t=30, b=20),
                  xaxis=dict(title="YEAR"), yaxis=dict(title="TOP rate"),
                  title="Average top rate (proxy for difficulty)",title_x=0.5, width=1000)
fig.show('png')
```

## Country medal tally

```python
medals = men.groupby("country").sum().sort_values("gold_B", ascending=False)[["gold_B","silver_B","bronze_B"]].head(10)
medals["total_medals_B"] = medals.sum(axis=1)
medals.rename(columns={col:' '.join(col.split('_')[:-1]).capitalize() for col in medals.columns}, inplace=True)
medals
```

```python
fig = px.bar(medals,y=["Gold", "Silver", "Bronze"], barmode="group",
             labels={"country": "Country", "value": "Medals"},
             color_discrete_sequence=[ifsc_pink,ifsc_blue,"#8affc1"],)
fig.update_layout(legend_title=None, title="Medal tallies by country",title_x=0.5, width=1000)
fig.show('png')
```
